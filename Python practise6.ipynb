1. What are escape characters, and how do you use them?
Ans: Backslash "\" is known as escape character. It is used for various reasons as mentioned below. <br>

Notation             use
\'              To allow single/double quote inside a string that is surrounded by single/double quote
                e.g. print('Dpn\'t miss today\'s class') ==> Dpn't miss today's class
                
\\              backslash
                e.g. print('"\\" is the backslash') ==> "\" is the backslash
                
\n              new line
                e.g. print('Hellow \nworld!') ==>    Hello
                                                  World!
                                                  
\t              tab
                e.g. print('Hellow \tworld!') ==>  Hellow     world
 

\b              backspace
                e.g. print("python is widely used in industriesd\b. Learn it.") ==> python is widely used in industries. Learn it.
