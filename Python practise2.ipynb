4. What are the values of the following expressions?

not (5 > 4)
(5 > 4) or (3 == 5)
not ((5 > 4) or (3 == 5))
(True and True) and (True == False)
(not False) or (not True)

not (5 > 4) => False 
(5 > 4) or (3 == 5) => True

(True and True) and (True == False) => False 
(not False) or (not True) => True
