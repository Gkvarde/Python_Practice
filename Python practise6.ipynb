1. What are escape characters, and how do you use them?
Ans: Backslash "\" is known as escape character. It is used for various reasons as mentioned below. <br>

\' ==> To allow single/double quote inside a string that is surrounded by single/double quote <br>
                e.g. print('Dpn\'t miss today\'s class') ==> Dpn't miss today's class <br>
                
\\ ==> backslash <br>
                e.g. print('"\\" is the backslash') ==> "\" is the backslash <br>
                
\n ==> new line <br>
                e.g. print('Hellow \nworld!') ==>    Hello <br>
                                                     World!<br> 
                                                  
\t ==> tab <br>
                e.g. print('Hellow \tworld!') ==>  Hellow     world <br>
 

\b ==> backspace <br>
                e.g. print("python is widely used in industriesd\b. Learn it.") ==> python is widely used in industries. Learn it. <br>
