10. In regular expressions, what is the difference between the + and * characters? <br>
Ans: + matches one or more whereas * matches zero or more <br>

11. What is the difference between {4} and {4,5} in regular expression? <br>
Ans: a{4} would match exactly 4 'a' characters, not even 3 whereas a{4,5} would match from 4 to 5 'a' characters. <br>

12. What do you mean by the \d, \w, and \s shorthand character classes signify in regular expressions? <br>
Ans: <br>
\d ==> Matches any decimal digit. It is equivalent to [0-9] <br>
\w ==> Matches characters considered alphanumeric in the ASCII character set; this is equivalent to [a-zA-Z0-9_] <br>
\s ==> Matches characters considered whitespace in the ASCII character set; this is equivalent to [ \t\n\r\f\v] <br>


13. What do means by \D, \W, and \S shorthand character classes signify in regular expressions? <br>
Ans: <br>
\D ==> Matches any character which is not a decimal digit. This is the opposite of \d. If the ASCII flag is used this becomes the equivalent of [^0-9] <br>
\W ==> Matches any character which is not a word character. This is the opposite of \w. If the ASCII flag is used this becomes the equivalent of [^a-zA-Z0-9_] <br>
\S ==> Matches any character which is not a whitespace character. This is the opposite of \s. If the ASCII flag is used this becomes the equivalent of [^ \t\n\r\f\v] <br>

14. What is the difference between .*? and .*? <br>
Ans: .* performs greedy match and .*? performs non-greedy (reluctant) match <br>
e.g. 20200000200 <br>
Using 2.*2 would give 202000002, e.g. re.findall(r'2.*2', '20200000200') ==> Output: ['202000002'] <br>
Using 2.*?2 would give 202 e.g. re.findall(r'2.'*'?2', '20200000200') ==> Output: ['202'] <br>

15. What is the syntax for matching both numbers and lowercase letters with a character class? <br>

import regex as re <br>
a = re.compile(r"[a-z0-9]") <br>
a.findall('abcABC102') <br>


16. What is the procedure for making a normal expression in regex case insensitive? <br>
Ans: We use re.IGNORECASE or re.I as the second argument in re.compile() to match case insensitive. <br>

17. What does the . character normally match? What does it match if re.DOTALL is passed as 2nd argument in re.compile()? <br>
Ans: The . character normally matches any character except the newline character. If re.DOTALL is passed as the second argument to re.compile(), then the dot will also match newline characters. <br>

18. If numRegex = re.compile(r'\d+'), what will numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen') return? <br>
Ans: 'X drummers, X pipers, five rings, X hens' <br>

19. What does passing re.VERBOSE as the 2nd argument to re.compile() allow to do? <br>
Ans: The re.VERBOSE argument allows you to add whitespace and comments to the string passed to re.compile(). <br>

20. How would you write a regex that match a number with comma for every three digits? It must match the given following: <br>

'42' <br>
'1,234' <br>
'6,368,745' <br>
but not the following: <br>
'12,34,567' (which has only two digits between the commas) <br>
'1234' (which lacks commas) <br>

Ans: re.compile(r'^\d{1,3}(,\d{3})*$') <br>


21. How would you write a regex that matches the full name of someone whose last name is Watanabe? You can assume that the first name that comes before it will always be one word that begins with a capital letter. The regex must match the following: <br>
'Haruto Watanabe' <br>
'Alice Watanabe' <br>
'RoboCop Watanabe' <br>
but not the following: <br>
'haruto Watanabe' (where the first name is not capitalized) <br>
'Mr. Watanabe' (where the preceding word has a nonletter character) <br>
'Watanabe' (which has no first name) <br>
'Haruto watanabe' (where Watanabe is not capitalized) <br>

Ans: re.compile(r'[A-Z][a-z]*\sWatanabe') <br>

22. How would you write a regex that matches a sentence where the first word is either Alice, Bob, or Carol; the second word is either eats, pets, or throws; the third word is apples, cats, or baseballs; and the sentence ends with a period? This regex should be case-insensitive. It must match the following: <br>
'Alice eats apples.' <br>
'Bob pets cats.' <br>
'Carol throws baseballs.' <br>
'Alice throws Apples.' <br>
'BOB EATS CATS.' <br>
but not the following: <br>
'RoboCop eats apples.' <br>
'ALICE THROWS FOOTBALLS.' <br>
'Carol eats 7 cats.' <br>

Ans: re.compile(r'(Alice|Bob|Carol)\s(eats|pets|throws)\s(apples|cats|baseballs)\.', re.IGNORECASE)
